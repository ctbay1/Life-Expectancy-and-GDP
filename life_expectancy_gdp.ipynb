```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./all_data.csv')
display(data.head())
```

```python
countries = data['Country'].unique()
```

```python
sub_idx = 1
plt.figure(figsize=(14, 7))
for country in countries:
    country_data = data[data['Country'] == country]
    ax = plt.subplot(2, 3, sub_idx)
    sub_idx += 1
    sns.scatterplot(x=country_data['Life expectancy at birth (years)'], y=country_data['GDP'],hue=country_data['Year'], palette='flare')
    if sub_idx == 2:
        plt.figlegend(loc='center right', title='Year')
    ax.get_legend().remove()
    
    plt.title('Country: %s' %country)
    plt.xlabel('Life expectancy at birth (years)')
    plt.ylabel('GDP')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
```

```python
sub_idx = 1
plt.figure(figsize=(14, 7))
for country in countries:
    country_data = data[data['Country'] == country]
    plt.subplot(2, 3, sub_idx)
    sub_idx += 1
    plt.plot(country_data['Year'], country_data['GDP'])
    plt.title('Country: %s' %country)
    plt.xlabel('Year')
    plt.ylabel('GDP')
    
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.suptitle('GDP by Country', fontsize=14, color='red')
plt.show()
```

```python
sub_idx = 1
plt.figure(figsize=(14, 7))
for country in countries:
    country_data = data[data['Country'] == country]
    plt.subplot(2, 3, sub_idx)
    sub_idx += 1
    plt.plot(country_data['Year'], country_data['Life expectancy at birth (years)'])

    plt.title('Country: %s, avg: %s' %(country, round(country_data['Life expectancy at birth (years)'].mean(), 1)))
    plt.xlabel('Year')
    plt.ylabel('Life expectancy')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.suptitle('Life expentacy by country', fontsize=14, color='red')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(x='Country', y='Life expectancy at birth (years)', data=data, hue='Country', legend=True)
plt.xticks(rotation=30)
plt.show()
```
